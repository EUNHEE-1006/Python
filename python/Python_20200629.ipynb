```python
data = {'name' : ['Jason', 'Molly', 'Tina', 'Jake', 'Amy'],
         'age' : [42, 52, 36, 24, 73],
         'preTestScore' : [4, 24, 31, 2, 3],
         'postTestScore' : [ 25, 94, 57, 62, 70]}
df = pd.DataFrame(data, columns = ['name', 'age', 'preTestScore', 'postTestScore'])
df
```

```python
print(df['age'].sum())
print(df['preTestScore'].mean())
print(df['preTestScore'].describe()) # 데이터 분포를 알고 싶을때
print(df['preTestScore'].cumsum())

print(df['preTestScore'].var())
print(df['preTestScore'].std())
print(df['preTestScore'].skew()) #데이터가 한쪽으로 너무 쏠려있을때, 정규분포로 작업하기 위한 함수, skew로 데이터이 치우침을 파악 
```
