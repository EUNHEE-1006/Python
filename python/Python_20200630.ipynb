```python
movieIdListURL="http://www.kobis.or.kr/kobisopenapi/webservice/rest/movie/searchMovieList.json?key=121663d4a25b471d32ba87b2a3556938&itemPerPage=40&openStartDt=2020"

"http://www.kobis.or.kr/kobisopenapi/webservice/rest/movie/searchMovieList.json?key=121663d4a25b471d32ba87b2a3556938&itemPerPage=40&openStartDt=2020"
```

```python
#해당 웹페이지에 접속
movieIdPage = urllib.request.urlopen(movieIdListURL)
```

```python
#movieData가 파이선에서 사용가능한 딕셔너리구조로 되어있음. 
movieIdData = json.loads(movieIdPage.read().decode("utf8"), encoding="utf8")
movieIdData
```

```python
#모든 정보를 담을 DF
movieIDDF = pd.DataFrame()

#1개 영화 정보들에 대해서 처리할 DF
movieIDDF_temp = pd.DataFrame()
movieIDDF_temp = movieIDDF_temp.append(
    {
        "movieCd" : "",
        "movieNm" : "",
        "movieNmEn" : "",
        "openDt" : "",
        "directorpeopleNm" : ""},
    ignore_index = True)

temp_movieCd_idx = movieIDDF_temp.columns.get_loc("movieCd")
temp_movieNm_idx = movieIDDF_temp.columns.get_loc("movieNm")
temp_movieNmEn_idx = movieIDDF_temp.columns.get_loc("movieNmEn")
temp_openDt_idx = movieIDDF_temp.columns.get_loc("openDt")
temp_director_idx = movieIDDF_temp.columns.get_loc("directorpeopleNm")

num = len(movieIdData["movieListResult"]["movieList"])
for i in range(0, num) :
    #movieIDDF_temp에 i번째 영화의 정보 담기
    movieIDDF_temp.iat[0, temp_movieCd_idx] = movieIdData["movieListResult"]["movieList"][i]["movieCd"]
    movieIDDF_temp.iat[0, temp_movieNm_idx] = movieIdData["movieListResult"]["movieList"][i]["movieNm"]
    movieIDDF_temp.iat[0, temp_movieNmEn_idx] = movieIdData["movieListResult"]["movieList"][i]["movieNmEn"]
    movieIDDF_temp.iat[0, temp_openDt_idx] = movieIdData["movieListResult"]["movieList"][i]["openDt"]
    
    if movieIdData["movieListResult"]["movieList"][i]["directors"] :
        movieIDDF_temp.iat[0, temp_director_idx] = movieIdData["movieListResult"]["movieList"][i]["directors"][0]["peopleNm"]
    else :
        pass
 
    #movieIDDF_temp에 있는 i번째 영화에 대한 정보를 movieIDDF에 추가
    movieIDDF = movieIDDF.append(movieIDDF_temp, ignore_index = True)
    
    #movieIDDF_temp에 담겨져 있는 정보들을 초기화 하는 부분 
    for j in range(0, len(movieIDDF_temp.columns)) :
        movieIDDF_temp.iat[0, j] = ""
movieIDDF
```

```python
movieIDDF['year'] = movieIDDF['openDt'].apply(lambda x: str(x)[0:4]) #str(x)는 문자열에만 sliching 이 먹으니깐
movieIDDF['month'] = movieIDDF['openDt'].apply(lambda x: str(x)[4:6])
movieIDDF['day'] = movieIDDF['openDt'].apply(lambda x: str(x)[6:8])
```

```python
movieIDDF
```

```python
movieIDDF = movieIDDF.sort_values(by="openDt")
```

```python
movieIDDF.set_index("movieCd")
```

```python
col_list=["openDt", "year", "month", "day", "movieNm", "movieNmEn", "directorpeopleNm"]
movieIDDF = movieIDDF[col_list] #반드시 업데이트
movieIDDF.head(5)
```
