```python
import urllib.request
from bs4 import BeautifulSoup #BeautifulSoup = 패키지가 아니라, 패키지 속의 모델이름. 
import pandas as pd
```

```python
sampleURL = "http://www.kobis.or.kr/kobisopenapi/webservice/rest/movie/searchMovieList.xml?key=121663d4a25b471d32ba87b2a3556938&itemPerPage=40&openStartDt=2020"
sampleURL
```

```python
res = urllib.request.urlopen(sampleURL)
print(res)
```

```python
####Json과 다른점 : 바로 던져버리면 돼, 굳이 읽어서 던질 이유가 없어. 
#html.parser :어떠한 형식으로 던질거냐. parser종류대로 해줘야 
soup = BeautifulSoup(res, "html.parser")
print(soup)
```

```python
#모두 소문자!! =>가져온 데이터 확인하고 여기 있는 걸로 사용하기 
print(soup.prettify())
```

```python
#모든 정보를 담을 DF
movieIDDF = pd.DataFrame()

#1개 영화 정보들에 대해서 처리할 DF
movieIDDF_temp = pd.DataFrame()
movieIDDF_temp = movieIDDF_temp.append(
    {
        "movieCd" : "",
        "movieNm" : "",
        "movieNmEn" : "",
        "openDt" : "",
        "directorpeopleNm" : ""},
    ignore_index = True)

temp_movieCd_idx = movieIDDF_temp.columns.get_loc("movieCd")
temp_movieNm_idx = movieIDDF_temp.columns.get_loc("movieNm")
temp_movieNmEn_idx = movieIDDF_temp.columns.get_loc("movieNmEn")
temp_openDt_idx = movieIDDF_temp.columns.get_loc("openDt")
temp_director_idx = movieIDDF_temp.columns.get_loc("directorpeopleNm")

num = len(soup.find_all("movie"))

for i in range(0, num) :
    tempinfo = soup.find_all("movie")[i]
    #movieIDDF_temp에 i번째 영화의 정보 담기
    movieIDDF_temp.iat[0, temp_movieCd_idx] = tempinfo.find("moviecd").text
    movieIDDF_temp.iat[0, temp_movieNm_idx] = tempinfo.find("movienm").text
    movieIDDF_temp.iat[0, temp_movieNmEn_idx] = tempinfo.find("movienmen").text
    movieIDDF_temp.iat[0, temp_openDt_idx] =tempinfo.find("opendt").text
    
    if tempinfo.find("directors").find_all("peoplenm") !=[] :
        movieIDDF_temp.iat[0, temp_director_idx] = tempinfo.find("peoplenm").text
    else :
        pass
 
    #movieIDDF_temp에 있는 i번째 영화에 대한 정보를 movieIDDF에 추가
    movieIDDF = movieIDDF.append(movieIDDF_temp, ignore_index = True)
    
    #movieIDDF_temp에 담겨져 있는 정보들을 초기화 하는 부분 
    for j in range(0, len(movieIDDF_temp.columns)) :
        movieIDDF_temp.iat[0, j] = ""
movieIDDF
```

```python
movieIDDF['year'] = movieIDDF['openDt'].apply(lambda x: str(x)[0:4]) #str(x)는 문자열에만 sliching 이 먹으니깐
movieIDDF['month'] = movieIDDF['openDt'].apply(lambda x: str(x)[4:6])
movieIDDF['day'] = movieIDDF['openDt'].apply(lambda x: str(x)[6:8])

movieIDDF
```

```python
movieIDDF = movieIDDF.sort_values(by="openDt")
movieIDDF
```

```python
movieIDDF = movieIDDF.set_index("movieCd")
col_list=["openDt", "year", "month", "day", "movieNm", "movieNmEn", "directorpeopleNm"]
movieIDDF = movieIDDF[col_list]
movieIDDF
```

```python
movieIDDF.index
```

```python
movieIDDF["typeNm"]=""
movieIDDF["actors"]=""
movieIDDF["cast"]=""
movieIDDF["actorsNum"]=""

i_typeNm = movieIDDF.columns.get_loc("typeNm")
i_actors = movieIDDF.columns.get_loc("actors")
i_cast = movieIDDF.columns.get_loc("cast")
i_actorsNum = movieIDDF.columns.get_loc("actorsNum")

num = len(movieIDDF)
for i in range(0, num) :
    moviecd = movieIDDF.index[i]
    tempURL = "http://www.kobis.or.kr/kobisopenapi/webservice/rest/movie/searchMovieInfo.xml?key=121663d4a25b471d32ba87b2a3556938&movieCd="+moviecd
    movieInfoPage1= urllib.request.urlopen(tempURL)
    movieInfoData1 = BeautifulSoup(movieInfoPage1, "html.parser")  
    
    movieIDDF.iat[i, i_typeNm] = movieInfoData1.find("typenm").text
    if movieInfoData1.find("actors").text != '' :
        #movieInfoData1.find("peoplenm") 이라 하면, directors에 있는 peoplenm이 먼저 나와서 데이터가 안맞아 => 다이렉트로 접근하는게 조심해야해
        movieIDDF.iat[i, i_actors] = movieInfoData1.find_all("actor")[0].find("peoplenm").text
        movieIDDF.iat[i, i_cast] =movieInfoData1.find_all("actor")[0].find("cast").text   
        movieIDDF.iat[i, i_actorsNum] = len(movieInfoData1.find("actor"))
    else : 
        pass   
movieIDDF
```
