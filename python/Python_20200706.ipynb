```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_excel("sampledata/11_sales-funnel.xlsx")
data.head()
```

```python
data["Status"] = data["Status"].astype("category") #astype : 알파벳 순서대로 정렬

data["Status"].cat.set_categories( #set_categories : 내가 원하는 순서대로 
    ["won", "pending", "presented", "declined"], inplace=True)
data.head()
```

```python
data.info()#Status 카테고리 값이 바뀌어져 있음을 확인가능 
```

```python
#Name을 가로줄 기준으로 하족, 데이터 확인 => 숫자인 컬럼들만 들어온다. 
pd.pivot_table(data, index=["Name", "Rep","Manager"])
```

```python
pd.pivot_table(data, 
               index=["Manager", "Rep"], #가로줄 세팅
               values=["Price"],#세로줄 세팅
               aggfunc=np.sum) #어떤 값을 넣을 것인가
```

```python
pd.pivot_table(data, 
               index=["Manager", "Rep"], #가로줄 세팅
               values=["Price"],#세로줄 세팅
               aggfunc=[np.mean, len]) #어떤 값을 넣을 것인가
```

```python
pd.pivot_table(data, 
               index=["Manager", "Rep"], #가로줄 세팅
               values=["Price", "Quantity"],#세로줄 세팅
               aggfunc=[np.sum], columns=["Product"], fill_value=0) #항목별로 더한값을 보고 싶을 때!!(=np.sum과 비교)
```

```python
pd.pivot_table(data, 
               index=["Manager", "Rep", "Product"], 
               values=["Price", "Quantity"],
               aggfunc=[np.sum, np.mean], fill_value=0, margins=True) #보고자 하는 사람마음대로 데이터를 핸들링!!!
                                                                        #margins = > all 
```

```python
table = pd.pivot_table(data, 
               index=["Manager", "Status"], 
               values=["Price", "Quantity"],
               aggfunc={"Price" : np.sum, "Quantity" :len}, #내가 원하는 벨류에 원하는 값을 넣는다!!
               columns=["Product"],
               fill_value=0, margins=True)
table
```

```python
table.query('Status==["won", "pending"]') # query : 조건검색 가능, 작은 따옴표!! 주의 
```

```python
#crosstab : 각 컬럼을 가로세로로 세팅해주어야 
pd.crosstab([data["Manager"], data["Rep"]], [data["Price"], data["Quantity"]],
           margins=True)
```

```python
pd.crosstab([data["Manager"], data["Rep"]], data["Product"],
            values = data["Price"],
            aggfunc=[len, np.mean]). fillna(0) #fill_value가 없다.!!
```

```python
# 기준 : month, item
# 보고자 하는 값 : duration, network_type, date
# duration => 합, network_type =>건수, date =>제일 처음 값 : 기능을 문자열로 해야 한다.!!!!!!
data.groupby(["month", "item"]).agg({"duration" : "sum", 
                                     "network_type" : "count", 
                                     "date" : "first"})
```

```python
pd.pivot_table(data,
               index=["month", "item"],
               values=["duration", "network_type", "date"],
               aggfunc={"duration" : "sum", "network_type" : "count", "date" : "first"}
                )
```
